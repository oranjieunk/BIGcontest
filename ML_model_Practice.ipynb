# 전처리

## Payment를 Binning & one-hot-encoding

```python
# train["payment1"] = train_pay['payment_amount'] == -0.149898
# train["payment2"] = (train["Fare"] >= 10) & (train["Fare"] < 50 )
# train["payment3"] = train["Fare"] >= 50
```

```python
def preprocessing(data,data_pay):
    payment_list = ["payment1","payment2","payment3","payment4","payment5","payment6","payment7","payment8"]
    week = 8
    size = data.shape[0]
    for i in range(0,week):
        data[payment_list[i]] = data_pay['payment_amount'].copy()
    for i in range(0,size):
        row_num = data.index[data_pay['acc_id'][i*8] == data['acc_id']]
        for cnt in range(0, week):
            column = payment_list[cnt]
            data[column][row_num] = data_pay['payment_amount'].iloc[i*8+cnt]
            print(cnt, i*8+cnt)
    data.head()
    return data
```

```python
# 참고 : https://gist.github.com/yong27/7869662
# http://www.racketracer.com/2016/07/06/pandas-in-parallel/
from multiprocessing import Pool
# 이전에 선언
# import numpy as np

def _apply_df(args):
    df, func, kwargs = args
    return df.apply(func, **kwargs)

def apply_by_multiprocessing(df, func, **kwargs):
    # 키워드 항목 중 workers 파라메터를 꺼냄
    workers = kwargs.pop('workers')
    # 위에서 가져온 workers 수로 프로세스 풀을 정의
    pool = Pool(processes=workers)
    # 실행할 함수와 데이터프레임을 워커의 수 만큼 나눠 작업
    result = pool.map(_apply_df, [(d, func, kwargs)
            for d in np.array_split(df, workers)])
    pool.close()
    # 작업 결과를 합쳐서 반환
    return pd.concat(list(result))
```

```python
clean_train_reviews = apply_by_multiprocessing(\
    train, preprocessing, workers=4)
```

```python
payment_list = ["payment1","payment2","payment3","payment4","payment5","payment6","payment7","payment8"]
week = 8
size = train.shape[0]
for i in range(0,week):
    train[payment_list[i]] = train_pay['payment_amount'].copy()
for i in range(0,size):
    row_num = train.index[train_pay['acc_id'][i*8] == train['acc_id']]
    for cnt in range(0, week):
        column = payment_list[cnt]
        train[column][row_num] = train_pay['payment_amount'].iloc[i*8+cnt]
        print(cnt, i*8+cnt)
train.head()
```

```python
# 자동으로 분류 나누기
for payment in payment_list:
    train[payment+'Bi'] = pd.cut(train[payment], 4)
```

```python
# 기계가 데이터를 이해할 수 있도록 
# 카테고리 데이터를 one-hot-encoding 해준다.
def dummy_data(data, columns):
    for column in columns:
        data = pd.concat([data, pd.get_dummies(data[column], prefix = column)], axis=1)
        data = data.drop(column, axis=1)
    return data


dummy_columns = ["payment1Bin","payment2Bin","payment3Bin","payment4Bin","payment5Bin","payment6Bin","payment7Bin","payment8Bin"]
train_dummy = dummy_data(train, dummy_columns)
# test_dummy = dummy_data(test, dummy_columns)

print('원핫인코딩 전 shape')
print(train.shape)
# print(test.shape)

print('get_dummies로 원핫인코딩 후 shape')
print(train_dummy.shape)
# print(test_dummy.shape)
```

## Random forest 학습

```python
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트 분류기를 사용
forest = RandomForestClassifier(
    n_estimators = 800, n_jobs = -1, random_state = 2018, max_depth = 10)
forest
```

```python
drop_columns = ['label','acc_id',"payment1","payment2","payment3","payment4","payment5","payment6","payment7","payment8"]
train_features = train.drop(columns=drop_columns)
test_features = test.drop(columns=drop_columns)
```

```python
# 테스트 데이터를 넣고 예측합니다.
y_pred = forest.predict(test_features)
y_pred[:3]
```
